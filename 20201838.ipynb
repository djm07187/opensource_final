```python
# Q.1 10점
#
# 문자열 my_string과 target이 매개변수로 주어질 때,
# target이 문자열 my_string의 부분 문자열이라면 1을,
# 아니라면 0을 return 하는 solution 함수를 작성하시오.
#
# 제한사항
# 1 ≤ my_string 의 길이 ≤ 100
# my_string 은 영소문자로만 이루어져 있습니다.
# 1 ≤ target 의 길이 ≤ 100
# target 은 영소문자로만 이루어져 있습니다.


def solution(my_string, target):
    
    answer = 0

    for i in range(len(my_string)):                     #my_string의 길이만큼 for문을 돌림
        if my_string[i] == target[0]:                   #target의 첫 번째 글자를 my_string에서 찾을 떄
            
            if my_string[i : i+len(target)] == target:  #my_string에서 target의 첫 번째 글자와 같은 위치(인덱스)에서 target의 길이만큼 슬라이싱하여 슬라이싱한 부분이 target이랑 같은지 비교한다.
                answer = 1                              #슬라이싱 한 부분이 target이랑 같으면 answer의 값을 1로 설정한다.
                break
        i += 1                                          #i의 크기를 1씩 증가시켜 my_string의 전부 다 확인한다.

    return answer


my_string = 'asdfasdfgghhhhhelloworld'
target = 'hello'
print(solution(my_string, target))


```

```python

# Q.2 10점
#
# 모스부호 해독 프로그램 만들기
# 문자열 letter 가 매개변수로 주어질 때,
# letter 영어 소문자로 바꾼 문자열을 return 하는
# solution 함수를 완성하시오.
#
# 제한사항
# 1 ≤ letter 의 길이 ≤ 1,000
# letter 의 모스부호는 공백으로 나누어져 있습니다.
# letter 에 공백은 연속으로 두 개 이상 존재하지 않습니다.
#
# letter = 여러분의 좌우명 또는 인상 깊었던 말을 쓰시오.

def solution(letter):
    morse = { 
    '.-':'a','-...':'b','-.-.':'c','-..':'d','.':'e','..-.':'f',
    '--.':'g','....':'h','..':'i','.---':'j','-.-':'k','.-..':'l',
    '--':'m','-.':'n','---':'o','.--.':'p','--.-':'q','.-.':'r',
    '...':'s','-':'t','..-':'u','...-':'v','.--':'w','-..-':'x',
    '-.--':'y','--..':'z'}

    answer = ''
    code = letter.split()                   #letter를 스페이스바 기준으로 split하여 code에 저장한다.

    for i in range(len(code)):              #code의 요소 개수만큼 for문을 돌린다.
       
        if code[i] == '/':                  #모스부호의 '/'는 스페이스바를 의미하기 때문에 code의 요소중에 '/'를 찾으면 answer의 값에 ' '를 더한다.
            answer += ' '
        else : 
            answer += morse.get(code[i])    #get함수를 사용하여 code에 있는 모스부호 값에 맞는 알파벳을 가져와서 answer에 더한다.
        
        i += 1                              #모든 code의 요소를 해독해야 함으로 1씩 더해준다.
        
    return answer


#life is choices between birth and death
letter = '.-.. .. ..-. . / .. ... / -.-. .... --- .. -.-. . ... / -... . - .-- . . -. / -... .. .-. - .... / .- -. -.. / -.. . .- - ....'
print(solution(letter))
```

```python
# Q.3 10점
#
# 행성의 나이를 알파벳으로 표현할 때,
# a는 0, b는 1, ..., j는 9
# 예를 들어 cd는 23살, fb는 51살입니다.
# age가 매개변수로 주어질 때
# PROGEAMMER-857식 나이를 return 하도록
# solution 함수를 완성하시오.
#
# 제한사항
# age는 자연수입니다.
# age ≤ 1,000
# PROGRAMMERS-857 행성은 알파벳 소문자만 사용합니다.

def solution(age):
    answer = ''
    abc = 'abcdefghij'          #abcdefghij 알파벳을 abc에 list로 저장
    
    change = str(age)           #age를 str로 이루어진 list(change)로 만들어서 각 자리의 숫자를 떨어지게 만듬 ex) 23 -> 2,3

    for i in change:            #i에다가 change의 각 요소를 집어넣는 for문을 돌림
        answer += abc[int(i)]   #각 요소값에 해당하는 알파벳이 answer에 저장되게 함 ex) 2 -> c , 3 -> d
        
    return answer


age = 23
print(solution(age)) 
```

```python
# Q.4 10점
#
# x축과 y축으로 이루어진 2차원 직교 좌표계에 중심이 원점인
# 서로 다른 크기의 원이 두 개 주어집니다.
# 반지름을 나타내는 두 정수 r1, r2가 매개변수로 주어질 때,
# 두 원 사이의 공간에 x좌표와 y좌표가 모두 정수인 점의 개수를
# return하도록 solution 함수를 완성해주세요.
# ※ 각 원 위의 점도 포함하여 셉니다.
#
# 제한사항
# 1 ≤ r1 < r2 ≤ 1,000,000

def solution(r1, r2):

    answer = 0                          #answer, r1_num, r2_num을 0으로 초기화한다.
    r1_num = 0                          #r1_num과 r2_num은 각각 r1과 r2가 만들어지는 원안에 존재하는 점의 개수를 저장하는 변수이다.
    r2_num = 0
    
    for x in range(-r2, r2 + 1):        #x값을 -r2부터 -r2+1까지 for문으로 돌린다. 이때 -r2+1인 이유는 마지막 부분까지 포함하기 위해서다.
        for y in range(-r2, r2 + 1):    #y값을 -r2부터 -r2+1까지 for문으로 돌린다. 이때 -r2+1인 이유는 마지막 부분까지 포함하기 위해서다.
            if x*x + y*y <= r1*r1:      #r1원 내부에 있는 점인지 확인하려면 (x,y)에서 원점까지의 거리가 원의 반지름보다 작거나 같으면 된다. 내부에 있는 점이면 if문이 작동되어 r1_num의 값이 1 증가한다.
                r1_num += 1

            if x*x + y*y <= r2*r2:      #r2원 내부에 있는 점인지 확인하려면 (x,y)에서 원점까지의 거리가 원의 반지름보다 작거나 같으면 된다. 내부에 있는 점이면 if문이 작동되어 r2_num의 값이 1 증가한다.
                r2_num += 1
            
            y += 1                      #x좌표를 고정하고 y좌표를 움직여서 모든 부분을 확인한다.
        x += 1
        
    answer = r2_num - r1_num            #r2가 r1보다 큼으로 answer의 값은 r2_num에서 r1_num을 뺀다.

    return answer

r1 = 0
r2 = 2
print(solution(r1,r2))

```

```python
# Q.5 10점
#
# 0 또는 양의 정수가 주어졌을 때,
# 정수를 이어 붙여 만들 수 있는 가장 큰 수를 알아내 주세요.
#
# 예를 들어, 주어진 정수가 [6, 10, 2]라면
# [6102, 6210, 1062, 1026, 2610, 2106]를 만들 수 있고,
# 이중 가장 큰 수는 6210입니다.
#
# 0 또는 양의 정수가 담긴 배열 numbers가 매개변수로 주어질 때,
# 순서를 재배치하여 만들 수 있는 가장 큰 수를 문자열로 바꾸어
# return 하도록 solution 함수를 작성해주세요.
#
# 제한사항
# numbers의 길이는 1 이상 100,000 이하입니다.
# numbers의 원소는 0 이상 1,000 이하입니다.
# 정답이 너무 클 수 있으니 문자열로 바꾸어 return 합니다.
#
# numbers = [8, 30, 17, 2, 23]

def solution(numbers):
    allList = [[]]                                  #이어 붙일 수 있는 모든 경우의 수를 저장하는 allList 초기화
    big = 0                                         #가장 큰 값을 저장하는 big을 0으로 초기화

    for num in numbers:                             #num안에 numbers의 요소를 순서대로 넣는 for 반복문
        new_allList = []                            #allList에 저장하기 전에 임시로 저장해 두는 new_allList 초기화
        for all in allList:                         #allList의 요소들 내부에 num을 넣어 모든 경우의 수를 만들어 주는 for문
            for i in range(len(all) + 1):           #all의 길이보다 1을 더 크게 한 이유는 index값과 len의 값의 차이 때문이다.
                new_all = all[:i] + [num] + all[i:] #요소 사이사이 num의 값을 넣기 위해서 만든 것(슬라이싱을 이용하여 i번째 사이에 num을 넣기 위해서는 all[:i] + all[i:] = all이기 때문에 두개 사이에 [num]을 더해주면 된다.)
                new_allList.append(new_all)         #append 함수를 사용하여 new_all을 new_allList에 저장한다.
                
        allList = new_allList                       #임시 저장한 new_allList를 allList에 저장한다.

    for all in allList:                             #모든 요소를 이어 붙여서 가장 큰 수를 찾는 for 반복문
        num = int(''.join(map(str, all)))           #all은 int형이기 때문에 이어 붙이기 위해서는 str로 바꾸어야한다. 그렇게 각각의 요소를 str로 바꾸고 join함수를 사용하여 하나로 이어 붙인다.
        if num > big:                               #big의 처음 값은 0 인데 처음 loop에서 num의 값이 더 크기 때문에 big의 값이 num의 값으로 바뀌고 모든 요소를 확인하면서 가장 큰 수를 big에 저장하게 된다.
            big = num
    return big

numbers = [697, 98, 283, 993]
print(solution(numbers))
```
